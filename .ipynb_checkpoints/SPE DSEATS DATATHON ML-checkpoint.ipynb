#  SPE DSEATS Africa Datathon 2025 - Well Classification

This notebook classifies 20 oil wells and estimates total oil production per reservoir using provided datasets.

## Objective
- Classify 20 wells by:
  - Reservoir Name
  - Reservoir Type (Saturated/Undersaturated)
  - Well Type (NF/GL)
  - Production Type (Steady/Unsteady)
  - GOR Trend (aSolGOR/bSolGOR/Combo)
  - Watercut Trend (Flat/Incr/Decr/Combo)
  - Oil PI Trend (Flat/Incr/Decr/Combo)
- Estimate total oil production (reservoir barrels) for five reservoirs (ACHI, KEMA, MAKO, DEPU, JANI)

## Workflow
- Load and explore datasets.
- EDA: Plot oil, gas, water production trends.
- Preprocess: Clean data, calculate rates, encode labels.
- Feature Engineering:
  - Assign reservoirs (200 psi rule).
  - Classify types and trends.
- Model: Train ML models (RandomForest, XGBoost, LogisticRegression).
- Estimate total oil production per reservoir.

```python
# Load Data and Explore Dataset

# Load datasets
wells_data = pd.read_csv("C:/Users/osuol/Desktop/SPE DSEAT/spe_africa_dseats_datathon_2025_wells_dataset.csv")
reservoir_info = pd.read_csv("C:/Users/osuol/Desktop/SPE DSEAT/reservoir_info.csv")

# Show wells data
wells_data.head()
```

### EDA - Production Trends

```python
# Check column names to avoid KeyError
wells_data.columns
```

```python
# Convert PROD_DATE to datetime
wells_data['PROD_DATE'] = pd.to_datetime(wells_data['PROD_DATE'], format='%d-%b-%y')
```

```python
# Matplotlib: Static plot to visualize (oil, gas, water) trends for Well_#1, Well_#5, Well_#10 
wells_to_plot = ['Well_#1', 'Well_#5', 'Well_#10']
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

for i, well in enumerate(wells_to_plot):
    well_data = wells_data[wells_data['WELL_NAME'] == well]
    axes[i].plot(well_data['PROD_DATE'], well_data['CUMULATIVE_OIL_PROD (STB)'], label='Oil (STB)', color='blue')
    axes[i].plot(well_data['PROD_DATE'], well_data['CUMULATIVE_FORMATION_GAS_PROD (MSCF)'], label='Gas (MSCF)', color='green')
    axes[i].plot(well_data['PROD_DATE'], well_data['CUMULATIVE_WATER_PROD (BBL)'], label='Water (BBL)', color='red')
    axes[i].set_title(f'Production Trends for {well}')
    axes[i].set_xlabel('Date')
    axes[i].grid(True)
    if i == 0:
        axes[i].set_ylabel('Cumulative Production')
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
for well in wells_data['WELL_NAME'].unique():
    well_data = wells_data[wells_data['WELL_NAME'] == well]
    plt.plot(well_data['PROD_DATE'], well_data['CUMULATIVE_OIL_PROD (STB)'], label=well)
plt.xlabel("Date", fontsize=13, labelpad=20)
plt.ylabel("Cumulative Oil (STB)", fontsize=13, labelpad=20)
plt.title("Cumulative Oil Production by Well", fontsize=20, pad=20, loc="left")
plt.grid(which="major", linestyle="-", alpha=0.3)
plt.minorticks_on()
plt.grid(which="minor", linestyle="--", alpha=0.1)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('cumulative_oil_all_wells.png', dpi=300)
plt.show()
```
